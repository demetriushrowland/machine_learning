```python
import numpy as np
import numpy.linalg as la
from scipy.integrate import dblquad
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
PI = np.pi
norm = 0


def prior(weights):
    return 1/np.sqrt(2*PI) * np.exp (-1/2*(np.dot(weights,weights)/10e4))

def likelihood (weights, data):
    n = len (data)
    p = len(data[0]) - 1
    targets = [row[-1] for row in data]
    X = [row[1:p+1] for row in data]
    product = 1
    for i in range(n):
        estimate = np.dot(X[i], weights)
        product *= 1/np.sqrt(2*PI) * np.exp (-1/2*(targets[i] - estimate)**2)
    return product

def posterior (weights, data):
    return prior (weights) * likelihood (weights, data)

def main():
    X = [np.random.uniform(-10,10) for time in range(100)]
    targets = [3*x + 5 for x in X]
    data = [[1] + [X[i]] + [targets[i]] for i in range(len(X))]
    
    weights_domain = []
    
    weight0 = -10
    while weight0 <= 10:
        weight1 = -10
        while weight1 <= 10:
            weights = [weight0, weight1]
            weights_domain.append (weights)
            weight1 += 1
        weight0 += 1
        
    max_weights = [0,0]
    max_posterior = 0
        
    for weights in weights_domain:
        ax.scatter (weights[0], weights[1], prior (weights))
        posterior_value = posterior(weights, data)
        if posterior_value > max_posterior:
            max_weights = weights
            max_posterior = posterior_value
        ax.scatter (weights[0], weights[1], posterior_value)
    
    print ('Max weights:', max_weights)
    
    
    
    plt.show()
            
    

main()
```
